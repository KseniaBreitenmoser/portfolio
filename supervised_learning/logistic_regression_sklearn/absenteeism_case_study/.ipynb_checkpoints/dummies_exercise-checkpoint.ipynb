```python
import pandas as pd
```

```python
raw_csv_data = pd.read_csv('Absenteeism_data.csv')
```

```python
df = raw_csv_data.copy()
```

```python
df = df.drop(['ID'], axis = 1)
```

```python
age_columns = pd.get_dummies(df['Age'])
```

```python
age_columns = age_columns.astype(int)
age_columns
```

```python
age_columns['check'] = age_columns.sum(axis = 1)
```

```python
age_columns
```

```python
age_columns['check'].sum(axis = 0)
```

```python
age_columns['check'].unique()
```

```python
age_columns = age_columns.drop(['check'], axis=1)
```

```python
df = df.drop(['Age'],axis = 1)
```

```python
df = pd.concat([df,age_columns], axis = 1)
```

```python
df.columns.values
```
