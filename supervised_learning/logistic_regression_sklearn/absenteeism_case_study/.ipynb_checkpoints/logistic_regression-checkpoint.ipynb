```python
import pandas as pd
import numpy as np
```

### Load the preprocessed data

```python
data_preprocessed = pd.read_csv('Absenteeism_preprocessed.csv')
```

```python
data_preprocessed.head()
```

### Create the targets

```python
# get the median of Absenteeism Time in Hours and use it as a cut off
# everything below - Moderate
# everything above - Excessive
data_preprocessed['Absenteeism Time in Hours'].median()
```

```python
targets = np.where(data_preprocessed['Absenteeism Time in Hours'] > 
                   data_preprocessed['Absenteeism Time in Hours'].median(), 1, 0)
```

```python
targets
```

```python
data_preprocessed['Excessive Absenteeism'] = targets
```

```python
data_preprocessed.head(10)
```

### Check data balance

```python
targets.sum() / targets.shape[0]
```

### Data with targets checkpoint

```python
# the columns droped are the ones that delivered weights close to zero and thus sebsequently removed
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours', 
                                            'Body Mass Index', 'Age',
                                           'Transportation Expense'], axis=1)
```

```python
data_with_targets is data_preprocessed
```

```python
data_with_targets.head()
```

### Select the inputs for the regression

```python
data_with_targets.shape
```

```python
data_with_targets.iloc[:,:-1]
```

```python
unscaled_inputs = data_with_targets.iloc[:,:-1]
unscaled_inputs
```

### Standardize the data

```python
from sklearn.preprocessing import StandardScaler

absenteeism_scaler = StandardScaler()
```

```python
dummies = unscaled_inputs[['reason_1', 'reason_2', 'reason_3', 'reason_4','Education']]
# to_scale = unscaled_inputs[['Month value',
#       'Day of the week', 'Distance to Work',
#       'Daily Work Load Average',
#       'Children', 'Pets']]
to_scale = unscaled_inputs[[x for x in unscaled_inputs.columns.values if x not in dummies]]
to_scale
```

```python
absenteeism_scaler.fit(to_scale)
```

```python
scaled_inputs_no_dummies = absenteeism_scaler.transform(to_scale)
scaled_inputs_no_dummies
```

```python
dummies = dummies.to_numpy()
dummies.shape[0]
```

```python
scaled_inputs_all = []
for i in range (dummies.shape[0]):
    scaled_inputs_all.append(np.concatenate((dummies[i],scaled_inputs_no_dummies[i])))
```

```python
scaled_inputs = np.array(scaled_inputs_all)
```

```python
scaled_inputs.shape
```

### Split the data into test and train and shuffle

#### Import the relevant module

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, train_size=0.8, random_state=20)
```

```python
print(x_train.shape, y_train.shape)
```

```python
print(x_test.shape, y_test.shape)
```

### Logistic Regression with SKLearn

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
```

```python
# machine learning
reg.fit(x_train, y_train)
```

```python
reg.score(x_train, y_train)
```

#### Manually check the accuracy

```python
model_outputs = reg.predict(x_train)
```

```python
model_outputs
```

```python
y_train
```

```python
model_outputs == y_train
```

```python
np.sum(model_outputs == y_train)
```

```python
model_outputs.shape[0]
```

```python
accuracy = np.sum(model_outputs == y_train) / model_outputs.shape[0]
accuracy
```

```python
print (reg.score(x_train, y_train), accuracy)
```

### Finding the intercept and coefficients

```python
reg.intercept_
```

```python
reg.coef_
```

```python
# get names of columns from unscaled_inputs (scaled_inputs are numpy arrays and dont' have names)
unscaled_inputs.columns.values
```

```python
feature_name = unscaled_inputs.columns.values
```

```python
summary_table = pd.DataFrame(columns=['Feature Name'], data = feature_name)
```

```python
summary_table['Coefficients'] = np.transpose(reg.coef_)
```

```python
summary_table
```

```python
# empty the 0th index in the summary_table to incert intercept 
summary_table.index = summary_table.index + 1
```

```python
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
```

```python
summary_table = summary_table.sort_index()
summary_table
```

```python
summary_table['Odds Ratio'] = np.exp(summary_table.Coefficients)
```

```python
summary_table.sort_values('Odds Ratio', ascending=False)
```

```python
# The features with coefficients close to zero can be neglected
# Interpret Reason 3: if the person reported Reason 3 (food poisoning) 
# the odds of excessive adbsence increase by 25 time as opposed to no reason
```

### Testing the model

```python
reg.score(x_test, y_test)
```

```python
y_hat = reg.predict(x_test)
```

```python
cm = metrics.confusion_matrix(y_test, y_hat)
print("Confusion Matrix:")
print(cm)
```

```python
metrics.ConfusionMatrixDisplay.from_predictions(y_test, y_hat)
```

```python
# Generate report
report = metrics.classification_report(y_test, y_hat)
print(report)
```
